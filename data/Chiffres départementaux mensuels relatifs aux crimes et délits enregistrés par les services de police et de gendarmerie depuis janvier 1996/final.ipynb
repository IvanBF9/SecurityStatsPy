On crée un dictionaire avec les années

```python
dict_year = {}
for col_name in chiffre_departement_df.columns:
    if '_' in col_name:
        s=col_name.split("_")
        year = s[1]
        month = s[2]
        if year in col_name:
            if year in dict_year:
                  dict_year[year].append(col_name)
            else:
                  dict_year[year] = [col_name]
dict_year
```

```python
for year in dict_year:
    chiffre_departement_df[year] = chiffre_departement_df[dict_year[year]].sum(axis=1)
```

```python
chiffre_departement_df = chiffre_departement_df.drop(['Index'], axis=1)
```

```python
for i in range(3,98):
    df = pd.read_excel('tableaux-4001-ts.xlsx', sheet_name=i)
    df = df.drop(['Index'], axis=1)
    df = df.set_index('libellé index')
    df = df.T
    if i < 21:
        departement = str(i - 1)
    elif i == 21:
        departement = '2A'
    elif i == 22:
        departement = '2B'
    elif i > 22:
        departement = str(i - 2)
    df['departement'] = departement
    df = df.reset_index()
    chiffre_departement_df = pd.concat([chiffre_departement_df, df], ignore_index=True)
```
