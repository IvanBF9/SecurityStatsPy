```python
filepath = 'tableaux-4001-ts.xlsx'
xl = pd.ExcelFile('tableaux-4001-ts.xlsx')
print(xl.sheet_names)
dict_df = pd.read_excel(filepath, xl.sheet_names)
```

```python
france_entiere_df = dict_df.get('France_Entière')
```

```python
france_entiere_df.shape
```

```python
france_entiere_df.columns
```

```python
dict_year = {}
for col_name in france_entiere_df.columns:
    if '_' in col_name:
        s=col_name.split("_")
        year = s[1]
        month = s[2]
        if year in col_name:
            if year in dict_year:
                  dict_year[year].append(col_name)
            else:
                  dict_year[year] = [col_name]
dict_year
```

```python
for year in dict_year:
    france_entiere_df[year] = france_entiere_df[dict_year[year]].sum(axis=1)
```

```python
for col_name in france_entiere_df.columns:
    if '_' in col_name:
        france_entiere_df = france_entiere_df.drop(col_name, axis=1)
```

```python
france_entiere_df = france_entiere_df.set_index('libellé index')
```

```python
france_entiere_df = france_entiere_df.drop('Index', axis=1)
```

```python
france_entiere_df = france_entiere_df.T
```
